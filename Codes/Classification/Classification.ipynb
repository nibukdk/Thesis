# Splitting Data

```python
#Import Train_Test_Split from  model_selection
from sklearn.model_selection import train_test_split
```

```python
# Splitting data into train and test independent and dependent variables respectively
X_train,X_test, y_train, y_test = train_test_split(df.iloc[:,:4], df.iloc[:,4], test_size=0.35, random_state=101)
```

```python
#Import StandardScaler for Scaling
#Note Scaling can also be done before splitting the data
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
# Scaling the data
X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)
```

# Train and Test Data with KNN

```python
# Import KNeaerestClassifier
from sklearn.neighbors import KNeighborsClassifier
```

```python
# Intialize KNN with 5 neighbors
knn = KNeighborsClassifier(n_neighbors=5)
```

```python
#Fit trains the data
knn.fit(X_train, y_train)
```

```python
y_pred = knn.predict(X_test)
```

# Evaluate Classification Model


```python
#Import Classification Report and Confusion Matrix
from sklearn.metrics import classification_report, confusion_matrix
```

```python
#Initialize Classification Report and Confusion Matrix
cls_report = classification_report(y_test,y_pred)
confusion_matrix = confusion_matrix(y_test,y_pred)
```

```python
print("Classification Report \n\n",cls_report)
print('\n')
print("Confusion matrix \n\n",confusion_matrix)
```

# Elobow Method
 Lets find the optimum value of k using elbow method.

```python
avg_error_rates = []
for i in range(1,11):
    knn= KNeighborsClassifier(i)
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)
    avg_error_rates.append(np.mean(y_test != y_pred))
```

```python
avg_error_rates
```

```python
plt.figure(figsize=(10,5))
plt.plot(range(1,11),avg_error_rates, c='b',ls='-', marker='o', markerfacecolor='y', markersize=10)
plt.title('Elbow Method Plot For Iris Data')
plt.xlabel('K Values')
plt.ylabel('Average Error Rates')
#plt.savefig('Elbow Method For K-Values')
```

# Grid Search CV


```python
from sklearn.model_selection import GridSearchCV
```

```python
#These are the parameters the Grid will iterate through to find the optimum value
params = {
    'n_neighbors':[1,2,3,5,5,6,7,8],
    'weights':['uniform', 'distance']
}
```

```python
#Initialize our GsCV
gridSearch = GridSearchCV(estimator=KNeighborsClassifier(), param_grid=params,verbose=5)
```

```python
gridSearch.fit(X_train,y_train)
```

```python
# Check the best Parmaters for our KNN after GSV
gridSearch.best_params_
```

So according to the params we passed. Grid Search thinks that value of k = 6 and weights is 'unform'

```python
gridSearch.best_estimator_
```
